```python
#Exercício 1

'''Faça um programa que peça ao usuário um número e imprima todos os números de um até o número dado.

Exemplo:

digite: 5

imprime: 1 2 3 4 5'''

num = int(input('Partindo de 1 até que número você quer que seja exibido? '))
i = 1
while i <= num:
    print(i)
    i += 1
```

```python
#Exercício 2

'''Peça ao usuário para digitar um número e imprima o fatorial de n.'''

num = int(input('Digite o número que deseja calcular seu fatorial. '))
i = 1
fat = 1

while i <= num:
    fat *= i
    i += 1

print(fat)
```

```python
#Exercício 3

'''Peça ao usuário para digitar um número N e some todos os números de 1 a N utilizando o laço de repetição while.'''

num = int(input('Qual o número que você deseja somar com seus antecessores? '))
i = 1
soma = 0

while i <= num:
    soma += i
    print(soma)
    i += 1
```

```python
#Exercício 4

'''Faça um programa que imprima a tabuada do 9 (de 9*1 a 9*10) usando loops.'''
i = 1

print('A tabuada do 9:\n')
while i <= 10:
    num = 9 * i
    print('9 x ', i, ' = ', num)
    i += 1
```

```python
#Exercício 5

'''Faça um programa que recebe um número inteiro do usuário e 
imprime na tela a quantidade de divisores desse número e quais são eles.'''

num = int(input('Qual o número que você gostaria de saber seus divisores e a quantidade deles? '))
num_div = 0
i = 1

print('Divisores de ', num, ':\n')

while i <= num:
    if num % i == 0:
        print(i)
        num_div += 1
    i += 1
    
print('\nNúmero de divisores que', num, 'tem:', num_div)
```

```python
#Exercício 6

'''Faça um programa, usando loops, que peça para um usuário digitar um número e que só finaliza
quando o usuário digitar 0. Ao final imprima a soma de todos os números digitados.'''
num = 1

while num != 0:
    num = float(input('Digite um número para ser exibido ou 0 para encerrar o programa. '))
    
    if num != 0:
        print(num)
        
    else:
        print('Programa encerrado.')
```

```python
#Exercício 7

'''Faça um programa que sorteia um número N e peça para o usuário adivinhar o número sorteado.
A cada resposta errada, o seu programa deve imprimir um aviso dizendo que a resposta está errada e
pedir novamente uma resposta ao usuário.

Para a realização desse exercício, utilize alguma função da biblioteca random (e.g. randint()).'''

import random
final = int(input('Partindo de 1 digite o intervalo que você quer que  um número seja sorteado. '))
num_sort = random.randint(1, final)
num = 0

print('Número sorteado (para verificação):', num_sort, '\n')

while num != num_sort:
    num = float(input('Digite seu chute. '))
    
    if num != num_sort:
        print('Errado, tente novamente!\n')
        
    else:
        print('Corrreto!')
```

```python
#Exercício 8

'''Faça um programa que peça para o usuário digitar a idade, o salário e o sexo de uma pessoa
até que as entradas digitadas sejam válidas.

a. Idade: entre 0 e 150;

b. Salário: maior que 0;

c. Sexo: M, F ou Outro.'''

idade = -1
salario = -1
sexo = 'a'

while idade < 0 or idade > 150:
    idade = int(input('Qual a idade do usuário? '))
    
    if idade < 0 or idade > 150:
        print('Idade inválida, por favor digite um valor de 0 a 150 anos.\n')
        
while salario < 0:
    salario = float(input('Qual o salário do usuário? '))
    
    if salario < 0:
        print('Salário inválido, por favor digite um valor maior que zero.\n')

while sexo != 'M' and sexo != 'F' and sexo != 'Outro':
    sexo = input('Qual o sexo do usuário? (M, F, Outro) ')
    
    if sexo != 'M' and sexo != 'F' and sexo != 'Outro':
        print('Sexo incorreto, por favor digite M para masculino, F para feminino ou Outro.\n')
        
print('\n', idade, '\n R$', salario, '\n', sexo)
```

```python
#Exercício 9

'''Desafio! - Calcule a soma de até mil termos da série 1 + 1/2 + 1/4 + 1/8 + 1/16 + ...

Dica: Use três variáveis:

• um contador, que começa em zero;

• uma variável para a soma de todos os termos, que também começa em zero;

• uma variável para cada termo, que começa em 1 e a cada loop é dividida por 2.

A repetição da soma de mil termos pode ser feita com a função while contador < 1000.'''

cont = 0
soma = 0
num = 1
final = int(input('Até que termo da série você quer realizar a soma? '))

while cont < final:
    soma += num
    print(soma)
    num /= 2
    cont += 1
```

```python
#Exercício 10

'''Super Desafio! - Calcule a soma de mil termos dos inversos dos fatoriais: 1/(1!) + 1/(2!) + 1/(3!) + 1/(4!) + ...

Dica: Assim como no exercício anterior use três variáveis: um contador; uma variável para a soma;
e uma variável para os termos. Lembre-se de que 4! = 4*3*2*1 que também é igual a 4*3!.'''
```

```python
cont = 0
soma = 0
final = int(input('Até que termo da série você que realizar a soma? '))


while cont < final:
    i = 1
    fat = 1
    
    while i <= (cont + 1):
        fat *= i        
        i += 1
        
    soma += 1 / fat
    print(soma)
    cont += 1    
```
