```python
#cria um dicionário com cpf como chave e perfil de investidor como valor
def cpf_profile(dados, posicao_cpf, posicao_profile):
    profile_dic = {}    
    for elem in dados:
        profile_dic[elem[posicao_cpf]] = elem[posicao_profile]
    return profile_dic

#calcula a distância euclidiana de duas listas com mesmo tamanho e retorna ela
def disteuc(lista1, lista2):
    dist = (sum([(x - y) ** 2 for x, y in zip(lista1, lista2)])) ** 0.5
        
    return dist

#classifica um ponto com base na quantidade de cada tipo dos K pontos mais próximos classificados
def classify(profiles, classificados_proximos):
    counter = 0
    profile_result = ''
    for profile in profiles:
        
        if counter < classificados_proximos.count(profile):
            counter = classificados_proximos.count(profile)
            profile_result = profile
    
    return profile_result
    
k = 5    
dist_dic = {}
estimated_profiles = {}

#Interagindo com um lista de cada vez referentes aos usuários não classificados e criando chaves no dicionário com seus cpf
for unclassified_client in no_class:
#unclassified_client = [[45926320819, '', (5800., 4000., 1200., 200.), ]] #PARA TESTAR
    dist_dic[unclassified_client[0]] = {}
    
    #Calculando a distância entre o usuário não classificado e todos os classificados → {cpf_unclass: {dist:cpf_class, ...}}
    for classified_client in data:        
        dist_dic[unclassified_client[0]][disteuc(unclassified_client[2], classified_client[2])] = classified_client[0]   
    
    #Alocar as distâncias e classificar elas do menor para o maior
    dist_only = list(dist_dic[unclassified_client[0]].keys())
    dist_only.sort()
    
    #Pegar os 5 menores entre eles
    closer_points = dist_only[:k]
    closer_classified = []
    
    #Dicionário CPF:Perfil dos usuários classificados
    classified_profiles = cpf_profile(data, 0, 1)
    
    #coloca na lista closer_classified o perfil dos clientes classificados mais próximos do não classificado
    for point in closer_points:
        closer_classified.append(classified_profiles.get(dist_dic[unclassified_client[0]].get(point)))

    #cria uma lista com todos os perfis existentes na lista sem repetir
    profiles = list(set(user[1] for user in data))
        
    #Dicionário CPF:Perfil dos usuários com o perfil estimado:
    estimated_profiles[unclassified_client[0]] = classify(profiles, closer_classified)     

print('\nestimated_profiles', estimated_profiles)
```
